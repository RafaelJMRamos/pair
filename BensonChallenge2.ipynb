```python
! curl http://web.mta.info/developers/data/nyct/turnstile/turnstile_160402.txt > 'mta.csv'

data = 'mta.csv'

import csv

def read_data(data):
  with open(data, 'r') as f:
    data =[row for row in csv.reader(f.read().splitlines())]
  return data

mta = read_data(data)

mta1 = [x[:4] for x in mta]

mta2 = [[x[6]+' '+x[7],x[9]] for x in mta]

import dateutil.parser

mta2 = mta2[1:]

mtatest2 = [[dateutil.parser.parse(x[0]), int(x[1])] for x in mta2]
```

```python
mta1 = mta1[1:]

mtatuple = []
for x in mta1:
    mtatuple.append(tuple(x))

from collections import defaultdict
mta2_dict =defaultdict(list)
for i,j in zip(mtatuple,mtatest2):
    mta2_dict[i].append(j)
```

```python
for key in mta2_dict: #loop through keys in dict
    for i in range(len(mta2_dict[key])-1): #loop through all values for each key
        mta2_dict[key][i][1] = abs(mta2_dict[key][i+1][1] - mta2_dict[key][i][1]) #key value gets difference of next time
    mta2_dict[key].pop() #remove the last date because no data
```
