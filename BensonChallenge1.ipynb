```python
! curl http://web.mta.info/developers/data/nyct/turnstile/turnstile_160402.txt > 'mta.csv'

data = 'mta.csv'

import csv

def read_data(data):
  with open(data, 'r') as f:
    data =[row for row in csv.reader(f.read().splitlines())]
  return data

mta = read_data(data)

mta1 = [x[:4] for x in mta]

mta2 = [x[4:] for x in mta]

for row in mta2:
    row[-1] = row[-1].strip()
```

```python
mta_dict = {}
```

```python
mtatupple = []
for row in mta1:
    mtatupple.append(tuple(row))
```

```python
from collections import defaultdict
mta_dict =defaultdict(list)
for i,j in zip(mtatupple,mta2):
    mta_dict[i].append(j)
```
